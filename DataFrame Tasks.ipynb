```python
import pandas as pd
dict1 = {
    "Name":["Parul","Harsh","Abhishek","Mayank"],
    "Fav_Color":["Black","Purple","Red","Blue"]
}
df = pd.DataFrame(dict1)
df
```

```python
dict2 = {
    "Name":["Priya","Harsh","Shruti","Mitesh"],
    "Age":[23,18,22,20]
}
df1 = pd.DataFrame(dict2)
df1
```

```python
pd.merge(df,df1,on = "Name",how= "outer")
```

```python
import pandas as pd
import numpy as np

area = np.random.randint(low = 2000,high = 5000,size = 20)
price = np.random.randint(low = 200000,high = 900000,size = 20)
rooms = np.random.randint(low = 10,high = 20,size = 20)
windows = np.random.randint(low = 20,high = 40,size = 20)   
dict1 = {
    "Area":area,
    "Price":price,
    "Rooms":rooms,
    "Windows":windows
    
}
df = pd.DataFrame(dict1)
df
```

```python
#saving it to a csv file
df.to_csv("price_data")
```
